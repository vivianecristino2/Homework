```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df_exemplo.csv',  delim_whitespace=True)
```

```python
#media
df['Salnty'].mean()
```

```python
#moda
df['Salnty'].mode()
```

```python
#mediana
df['Salnty'].median()
```

```python
#percentis

pct25 = df['Salnty'].quantile(0.25)

pct50 = df['Salnty'].quantile(0.50)

pct75 = df['Salnty'].quantile(0.75)

pct100 = df['Salnty'].quantile(1.00)

print('pct25:', pct25, 'pct50:', pct50, 'pct75:', pct75, 'pct100:', pct100)
```
